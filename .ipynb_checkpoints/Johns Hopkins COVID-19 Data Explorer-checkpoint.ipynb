## Source

The data used in this notebook is sourced from Johns Hopkins University's GitHub repository: https://github.com/CSSEGISandData/COVID-19. The data includes both confirmed and probable cases/deaths.

## Field Descriptions
* `FIPS` - US only. Federal Information Processing Standards code that uniquely identifies counties within the USA.
* `Lat` and `Long_` - Dot locations on the dashboard. All points (except for Australia) shown on the map are based on geographic centroids, and are not representative of a specific address, building or any location at a spatial scale finer than a province/state. Australian dots are located at the centroid of the largest city in each state.
* `UID` - Unique Identifier for each row entry.
* `Combined_Key`
* `iso2` and `iso3` - Officialy assigned country code identifiers.
* `Country_Region` - Country, region or sovereignty name. The names of locations included on the Website correspond with the official designations used by the U.S. Department of State.
* `code3` - 
* `Admin2` - county
* `Province_State` - Province, state or dependency name.

# Fetch and format data

```python
# load libraries
import pandas as pd, numpy as np, matplotlib.pyplot as plt,seaborn as sns,math,warnings
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
%config IPCompleter.greedy=True
%config IPCompleter.use_jedi=False
warnings.filterwarnings('ignore')

# urls for jhu datasets
cases_csv_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv'
deaths_csv_url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_US.csv'

# load dataframes from urls
cases_df = pd.read_csv(cases_csv_url)
deaths_df = pd.read_csv(deaths_csv_url)

# combine dfs with new column for report type
cases_df['Report_Type'] = 'cases'
deaths_df['Report_Type'] = 'deaths'
reports_df = cases_df.append(deaths_df,ignore_index=True)

# rename date columns (pad month and day values) for later filtering/sorting
date_cols = reports_df.filter(regex=r'\d{1,2}\/\d{1,2}\/\d{2}').columns.to_list()
date_formatter = lambda x: "{:0>2}/{:0>2}/{:0>2}".format(*(x.split('/')))
date_cols_map = {k: date_formatter(k) for k in date_cols}
reports_df = reports_df.rename(columns=date_cols_map)
date_cols = list(date_cols_map.values())
today = date_cols[-1]
```

# Filter data

```python
# list US states and District of Columbia for later filtering/sorting
us_states = reports_df[(reports_df['iso3']=='USA')& \
            (reports_df['Admin2'].notna())]['Province_State'].unique()

# create df for US states data
states_df = reports_df[reports_df['Province_State'].isin(us_states)] \
            .set_index(['Report_Type','Province_State','Admin2'])[date_cols]
```

# Explore data

## Functions for exploring and plotting

```python
# calculate a number of days based on a period of time (as a tuple)

def period_to_days(count,period_type='d'):
    """Convert a time period to a days value.
    
    Accepted `period_type` values are:
     - 'd' for day/s
     - 'w' for week/s
     - 'm' for months/s

    Default `period_type` is days.

    """
    if period_type == 'w':
        count *= 7
    elif period_type == 'm':
        count *= 30
    return count
```

```python
# filter frame by report type, state, county

def filter_df(state=None,county=None,report_type='cases',date=None,period=None,df=states_df):
    """Filter a dataframe based on a state, county, report type, date/period, or a combination.
    """
    if county:
        df = df.loc[(report_type,state,county)]#.to_frame(name=county).transpose()
    elif state:
        df = states_df.loc[(report_type,state)].sum()#.to_frame(name=state).transpose()
    else:
        df = states_df.loc[report_type].sum()#.to_frame(name='United States').transpose()
        
    if date:
        return df[date]
    elif period:
        return df[-period_to_days(*period):]
    else:
        return df
    
    
```

```python
# print report data in a consice statement

def print_report_data(report_count,state=None,county=None,report_type='cases',date=None,period=None):
    """Print report totals in a more readable format."""
    location = f"{county} County, {state}" if county else (state if state else "The United States")
    if period:
        count,period_type = period
        period_types = {
            'd':'days',
            'w':'weeks',
            'm':'months',
        }
        formatted_period = period_types[period_type]
        time = f"in the past {count} {formatted_period if count > 1 else formatted_period[:-1]}"
    else:
        time = f"on {date}"

    return f"""{location} reported {report_count} {report_type} {time}."""
    
print_report_data(state="Michigan",report_count=1000, period=(1,'w'))
```

```python
# calculate sum of reports over a period of time

def calculate_change(state=None,county=None,report_type='cases',period=None,df=states_df,**kwargs):
    """Calculate change over time for a state or county based on report type and time period.
    """
    df = filter_df(state=state,county=county,report_type=report_type,df=df)
    if period:
        adj_count = period_to_days(*period)+1
        df = df[-adj_count:]
        total = df.diff()[1:].sum()
    else:
        total = df.diff().sum()
        
    if kwargs.get('print_report_data',False)==True:
        return print_report_data(report_count=total,state=state,county=county,report_type=report_type,period=period)
    else:
        return total

print(calculate_change(state='Massachusetts',county='Norfolk',period=(1,'d'),print_report_data=True))
```

```python
# get cumulateive case and death counts

def cumulative_case_and_death_counts(state=False,county=False,date=date_cols[-1]):
    """Returns the total case and death counts for
    a given county and state as of a given date.
    If no state or county is passed, the US totals will be used
    If a date is not passed, the most recent date will be used."""
    
    if county:
        counts = states_df.query(f'Province_State=="{state}" & Admin2=="{county}"')[date]
    elif state:
        counts = states_df.query(f'Province_State=="{state}"').groupby(['Report_Type','Province_State']).sum()[date]
    else:
        counts = states_df.groupby('Report_Type').sum()[date].to_frame()
    
    res = f"""There have been {counts.loc['cases'][0]} cases of COVID-19 and """ \
            f"""{counts.loc['deaths'][0]} deaths reported in {f'{county} County, ' if county else ''}""" \
            f"""{state if state else 'the United States'} as of {date}."""
    
    return res
```

```python
# get first reported case date and return dates for plotting

def get_plot_dates(state=False):
    first_reported_grouping = ('cases',state) if state else ('cases')
    first_reported_case = states_df.loc[first_reported_grouping].sum().ne(0).idxmax()
    plot_dates = date_cols[date_cols.index(first_reported_case)-1:] if state else date_cols
    return plot_dates
```

```python
# plot daily change in cases/deaths for a given state or county

def plot_daily_change(state=False,county=False,report_type='cases',time_frame='all'):
    """Plots the daily change in cases/deaths for a given state or county.
If no state or county is passed, the total US count will be used.

Accepted `report_type` values are:
 - 'cases': case data
 - 'deaths': death data
    
"""
    plot_dates = get_plot_dates(state)
    
    if county:
        plot_df = states_df.loc[(report_type,state,county)].diff()[plot_dates]
    elif state:
        plot_df = states_df.loc[(report_type,state)].sum().diff()[plot_dates]
    else:
        plot_df = states_df.loc[report_type].sum().diff()[plot_dates]

    plt.figure(figsize=(12, 5))
    
    color = 'tab:red' if report_type=='cases' else 'tab:gray'
    
    daily_cases = plot_df.plot(kind='bar',color=color,alpha=.4)
    
    average_cases = plot_df.rolling(7,center=True).mean()
    average_cases.plot(linewidth=2,color=color)
    
    plt.xticks(ticks=np.arange(0,len(plot_dates), 15),labels=plot_dates[::15], \
                rotation=45,ha='right')
    
    plt.title(f"""Seven-day rolling average of daily COVID-19 {report_type} for """ \
              f"""{f'{county} County, ' if county else ''}""" \
              f"""{state if state else 'the United States'} as of {date_cols[-1]}""")
    
    plt.ylabel(f'Number of {report_type}')
    plt.xlabel('Date')
    plt.legend().remove()
    
    for i in list(plt.yticks()[0][::2]):
        plt.axhline(y=i,linewidth=1.25, color='lightgray',alpha=.75, \
                    linestyle='dashed',zorder=-5)
   
    plt.fill_between(plot_dates,average_cases,color=color,alpha=.15)
    plt.show()
```

```python
# view a report of cases/deaths with charts for a given state or county

def status_report(state=None,county=None): 
    

    print(cumulative_case_and_death_counts(state,county))

    for report_type in ['cases','deaths']:
        print(f"""
New {report_type} in past...
    1 day:\t{calculate_change(state,county,period=(1,'d'),report_type=report_type):.0f}
    7 days:\t{calculate_change(state,county,period=(7,'d'),report_type=report_type):.0f}
    14 days:\t{calculate_change(state,county,period=(14,'d'),report_type=report_type):.0f}

    Daily {report_type} average for past 7 days: {calculate_change(state,county,period=(7,'d'),report_type=report_type)/7:.0f}
        """)
        plot_daily_change(state,county,report_type=report_type)
```

# Experimental visualizations

```python
#average cases reported by day of the week (for the US or for a given state)
weekdays = {
    0:'0 Monday',
    1:'1 Tuesday',
    2:'2 Wednesday',
    3:'3 Thursday',
    4:'4 Friday',
    5:'5 Saturday',
    6:'6 Sunday'
}
state = None
report_type = 'cases'

if state:
    weekly_cases = states_df.loc[report_type].loc[state].sum().to_frame(name='cases')
else:
    weekly_cases = states_df.loc[report_type].sum().to_frame(name='cases')
weekly_cases.index.rename('date', inplace=True)
weekly_cases = weekly_cases.reset_index()
weekly_cases['date'] = pd.to_datetime(weekly_cases['date'])
weekly_cases['weekday'] = weekly_cases['date'].apply(lambda x: weekdays[x.weekday()])
weekly_cases['added_cases'] = weekly_cases['cases'].diff()
weekly_cases = weekly_cases[weekly_cases['date']>='2020-05-01']
cases_by_weekday = weekly_cases.groupby('weekday')['added_cases'].sum()/weekly_cases['weekday'].value_counts()
cases_by_weekday.plot.bar()
plt.show()
```

```python
# projections for daily and weekly cases based on a given daily report
state = 'Massachusetts'
cases_reported = 540
day_reported = '3 Thursday'

projected_cases = (cases_by_weekday/cases_by_weekday.sum()) \
    /(cases_by_weekday[day_reported]/cases_by_weekday.sum()) * cases_reported

print(f"COVID-19 case predictions for {state}",end='\n\n')
print("Projected cases per day:",projected_cases,sep='\n',end='\n\n')
print("Projected cases for the week:",projected_cases.sum(),sep='\n',end='\n\n')
```

```python
# cases in the last 7 days
weekly_change_df = states_df.loc['cases'].groupby('Province_State').sum().diff(axis=1,periods=7)[today].to_frame(name='new_cases')
weekly_change_df['total_cases'] = states_df.groupby('Province_State').sum()[today]
weekly_change_df['pct_change'] = weekly_change_df['new_cases']/weekly_change_df['total_cases']

cm = sns.light_palette("red", as_cmap=True)
weekly_change_df.sort_values(by='pct_change', ascending=False)#.style.background_gradient(cmap=cm)
```

```python
# plot an average line

report_type = 'cases'
state = 'Michigan'
days = 21

week_dates = date_cols[(-days-1):-1]
trend_df = states_df.query(f'Report_Type=="{report_type}" & Province_State=="{state}"').groupby(level='Province_State').sum().diff(axis=1)[week_dates]
# trend_df = trend_df[week_dates].where(~(trend_df[week_dates]<0), other=np.nan)
trend_df = trend_df.transpose().reset_index().rename(columns={'index':'Date'})
# print(trend_df)
x = trend_df.index
y = trend_df[state]
plt.bar(x,y,color='tab:gray',alpha=.75)
plt.xticks(ticks=np.arange(0,len(week_dates),2),labels=week_dates[::2],rotation=45,ha='right')
# trend line plot code from https://widu.tumblr.com/post/43624347354/matplotlib-trendline
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
avg_points = p(x)
avg_col = 'tab:green' if (avg_points[0]-  avg_points[1]) > 0 else 'tab:red'
plt.plot(x,avg_points,avg_col,linewidth=5)
plt.title(f'{days}-day trend for daily new {report_type} in {state}')
plt.show()
```

```python
# calculate totals, change, and percent change based on weekly interval
interval= 7
start = ((len(date_cols) - 1) % interval)
weekly_change_df = states_df.loc['cases'].groupby('Province_State').sum().diff(axis=1,periods=7)[date_cols[start::interval]] # weekly change ending on date
weekly_totals_df = states_df.loc['cases'].groupby('Province_State').sum()[date_cols[start::interval]] # case totals ending on date
weekly_change_pct_df = (weekly_change_df/weekly_totals_df)*100 # percent of weekly change over total cases

# cm = sns.light_palette("red", as_cmap=True)
# weekly_change_pct_df[weekly_change_pct_df.columns[-10:]].style.background_gradient(cmap=cm).highlight_null('white') # change in weekly percentage from week to week

cm = sns.diverging_palette(120,10, as_cmap=True)
weekly_change_pct_df.diff(axis=1)[weekly_change_pct_df.columns[-8:]].style.background_gradient(cmap=cm).highlight_null('white') # change in weekly percentage from week to week
```

```python
def plot_county_heatmaps(state,report_type='cases'):
    test_df = states_df.loc[(report_type,state)][date_cols].sort_values(by=[date_cols[-1]],ascending=False).diff(axis=1).fillna(0).head(10)
    test_df = test_df[date_cols].where(~(test_df[date_cols]<0), other=np.nan)
    counties = test_df.index.to_list()
    fig, ax = plt.subplots(len(counties)+1,1,figsize=(12, len(counties)/1.5))
    for i,county in enumerate(counties):
        nbins = 4
        bin_labels = [i for i in range(nbins)]
        plot = pd.cut(test_df.loc[county],bins=nbins,labels=bin_labels).astype('float').to_frame(name='cases').transpose()
        sns.heatmap(plot,cmap='Reds',cbar=False,yticklabels=[county],xticklabels=False,ax=ax[i])
        plt.setp(ax[i].yaxis.get_majorticklabels(), rotation=0)
    sns.heatmap(pd.DataFrame([1,2,3,4]).transpose(),cmap='Reds',cbar=False,ax=ax[-1],yticklabels=False, xticklabels=['Fewer','','','More'])
    ax[0].title.set_text(f'Daily Reported Cases in {state} - Top 10 Counties with Highest Total Cases\n') 
    plt.show()
    
plot_county_heatmaps('Michigan')
```

```python
def plot_state_heatmaps(report_type='cases',n_states=10):
    daily_cases_df = states_df.loc[report_type].groupby('Province_State').sum()[date_cols]
    daily_cases_df = daily_cases_df.sort_values(by=[date_cols[-1]],ascending=False).diff(axis=1).head(n_states)
    daily_cases_df = daily_cases_df[date_cols].where(~(daily_cases_df[date_cols]<0), other=np.nan)
    states = daily_cases_df.index.to_list()
    fig, ax = plt.subplots(len(states)+1,1,figsize=(12, len(states)/2))
    for i,state in enumerate(states):
        nbins = 4
        bin_labels = [i for i in range(nbins)]
        plot = pd.cut(daily_cases_df.loc[state],bins=nbins,labels=bin_labels).astype('float').to_frame().transpose()
        sns.heatmap(plot,cmap='Reds',vmin=-.5,cbar=False,yticklabels=[state],xticklabels=False,ax=ax[i])
        plt.setp(ax[i].yaxis.get_majorticklabels(), rotation=0)
    cbar = pd.DataFrame(range(nbins)).transpose()
    sns.heatmap(cbar,cmap='Reds',vmin=-.5,cbar=False,ax=ax[-1],yticklabels=False, xticklabels=['Fewer','','','More'])
    ax[0].title.set_text(f'Daily Reported COVID-19 {report_type}'
                         f' - Top {n_states} States with Highest Total {report_type}'
                         f' - {date_cols[0]} through {date_cols[-1]}\n') 
    plt.show()
    
plot_state_heatmaps(n_states=15)
```
